```python
folder_path="/content/"
text_files=[f for f in os.listdir(folder_path) if f.startswith("IV")]
text_files
```

```python
import os
import pandas as pd
from io import StringIO

csv_file = pd.DataFrame()

# Loop over all files
for file in text_files:
    list1 = []

    file_path = os.path.join(folder_path, file)
    with open(file_path, 'r') as f:
        lines = f.readlines()

    # Extract required lines and filter those with >4 elements
    for line in lines:
        if len(line.split()) > 4:
            list1.append(line.strip())

    # Join as string and read with pandas
    data_str = "\n".join(list1)
    temp_df = pd.read_csv(StringIO(data_str), sep="\t", header=None)

    # OPTIONAL: drop unwanted columns if known, e.g., drop columns 1, 2, 3
    # Make sure the index numbers are valid for temp_df
    temp_df.drop(columns=temp_df.iloc[:,[-1,-3,-6,-7,-8,-9]],inplace=True)

    # Reset column index if needed and concatenate
    temp_df.reset_index(drop=True, inplace=True)
    csv_file = pd.concat([csv_file, temp_df], axis=1)

# Done. Now csv_file contains cleaned and concatenated data
csv_file
```

```python
csv_file.drop(1,inplace=True)
csv_file
```
