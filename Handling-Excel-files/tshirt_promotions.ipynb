```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

bis=pd.read_excel("T shirt for pramotions.xlsx",names=["SN","Institute_name","Institute_cordinator","contact_number","Size_s","Size_M","Size_l","Size_xl","Size_xxl","Total"])
bis
```

```python
bis.drop(index=[0],inplace=True)
bis
```

```python
plt.figure(figsize=(10,8),dpi=80)
plt.bar(bis["Institute_name"],bis["Total"],edgecolor="r",color="c")
plt.xticks(fontsize=10,rotation=90)
for index,height in enumerate(bis["Total"]):
    plt.text(index,height,int(height),ha='center',va='bottom',size=16)
plt.ylabel("No of T-shirts",fontsize=15)
plt.title("T-shirts for promotion",fontsize=20)
plt.savefig("Bargraph tshirt.jpg",bbox_inches='tight')
plt.show()
```

```python

x=bis["Total"].values
plt.pie(x,labels=bis["Institute_name"],shadow=True,radius=1.6,rotatelabels=2,textprops={"fontsize":15})
plt.pie(x,labels=bis["Total"],shadow=True,radius=1.6,rotatelabels=2,textprops={"fontsize":15},labeldistance=0.7)
plt.savefig("Piegraph tshirt.jpg",bbox_inches='tight')
plt.show()
```

```python
size_no=[]
for i in range(4,9):
    size_no.append(bis[bis.columns[i]].sum())
list_name=["Size_s","Size_M","Size_L","Size_XL","Size_XXL"]
```

```python
bis["Total"].sum()
```

```python
plt.figure(figsize=(10,8))
plt.bar(list_name,size_no,edgecolor="r",color="c")
plt.xticks(fontsize=10,rotation=90)
for index,height in enumerate(size_no):
    plt.text(index,height,int(height),ha='center',va='bottom',size=16)
plt.ylabel("No of sizes",fontsize=15)
plt.title("Size distribution",fontsize=20)
plt.text(4,55,"Total=136",fontsize=10)
plt.savefig("Bargraph sizes.jpg",bbox_inches='tight')
plt.show()
```
